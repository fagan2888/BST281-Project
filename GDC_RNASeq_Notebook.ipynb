### Step 3: Load Data (RNA-Seq HT-Seq Counts) and format it for analysis

```python
# Copy the location of where the RNASeq data resides
import os

for Folder in os.listdir('../../../data/merged_rna_full/'):
    
    Data_Folder = '../../../data/merged_rna_full/' + Folder
    print(Folder.split('_')[-1])
    Data = pd.read_csv(Data_Folder + '/Merged_Counts.tsv',sep='\t')

    # Format Data:
    Data.drop(['gene_name'], axis=1, inplace=True)
    Data.rename(columns={ Data.columns[0]: "gene_name" }, inplace=True)
    Data.set_index('gene_name',inplace=True)
    Data = Data.iloc[:-5]

    # Select only protein coding genes
    with open('gdc-rnaseq-tool/EnsembleIDsPCG.txt') as f:
        PCGs = f.read().splitlines()
    Data = Data[Data.index.isin(PCGs)]
    Data.to_csv('../../../data/merged_rna/merged_rna_' + Folder.split('_')[-1] + '.tsv', sep='\t')
```

### Step 4: Load Clinical Metadata

```python
Metadata = pd.read_csv('metadata.files.txt',sep='\t')
Metadata.set_index('cases.0.samples.0.portions.0.analytes.0.aliquots.0.submitter_id',inplace=True)
Metadata
```

### Step 5: Perform Differential Expression analysis using Limma R Package

##### The central idea is to fit a linear model to the expression data for each gene.  See this document for more information https://kasperdanielhansen.github.io/genbioconductor/html/limma.html
##### Script for performing gene expression analysis located in the R script entitiled signature.R
##### Documentation for Limma located here: http://bioconductor.org/packages/release/bioc/html/limma.html

```python
# Organize Data for R Script
A_Name = 'Primary Tumor'
B_Name = 'Solid Tissue Normal'

group_A = list(Metadata[Metadata['cases.0.samples.0.sample_type'] == A_Name].index)
group_B = list(Metadata[Metadata['cases.0.samples.0.sample_type'] == B_Name].index)

Design_Matrix = pd.DataFrame([{'index': x, 'A': int(x in group_A), 'B': int(x in group_B)} for x in Data.columns]).set_index('index')
processed_data = {'expression': Data, 'design': Design_Matrix}

pd.DataFrame({A_Name:len(group_A),B_Name:len(group_B)},index=['Samples'])
```

```python
# Step 5: Run Limma R Script for calculating Log Fold Change

# Load the external R script for differential expression analysis
from rpy2.robjects import r, pandas2ri
r.source('signature.R')

# Run Data through R Script
Limma_Data = pandas2ri.ri2py(r.limma(pandas2ri.py2ri(processed_data['expression']), 
                     pandas2ri.py2ri(processed_data['design'])))
Limma_Data
```
